```python
import pandas as pd
import json
f = open('data.json', 'wb')

df = pd.read_csv('data.csv')

i = 0
j = 0
lst = []
menu = []
restaurant =  ""
address =  ""
category =  ""
rating =  ""
image =  ""
link =  ""
with open('d.json', 'w') as f:
    for r in df.iterrows():
        j += 1
        row = r[1]
        if row["LINK"] != link or link == "":
            print(row[0])
            if link != "":
                lst.append({
                            "id": i,
                            "restaurant":  restaurant,
                            "address": address,
                            "category":  category,
                            "rating":  rating,
                            "image": image,
                            "link":  link,
                            "menu": menu})
            i += 1
            restaurant =  row["RESTAURANT"]
            address =  row["ADDRESS"]
            category =  row["CATEGORY"]
            rating =  row["RATING"],
            image =  row["RESTAURANT IMAGE"]
            link =  row["LINK"]
            if len(menu) > 0: menu = []
        
        menu.append({"item_id": j,
                    "item": row["ITEM"],
                    "item_image": row["ITEM IMGAGE"],
                    "price": row["PRICE"]})
    lst.append({
                "restaurant":  restaurant,
                "address": address,
                "category":  category,
                "rating":  rating,
                "image": image,
                "link":  link,
                "menu": menu})
    json.dump(lst, f, indent = 4,ensure_ascii=False)
```
